# Performance Comparison

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('evaluation_results.csv')

models = df['MODEL']
iou = df['IOU']
dice = df['Dice']
pa = df['PA']
val_loss = df['VAL_LOSS']

def add_bar_values(ax, data):
    for i, val in enumerate(data):
        ax.text(i, val + 0.01, f'{val:.3f}', ha='center', va='bottom')
        
colors_iou = plt.cm.Blues(iou / iou.max())
colors_dice = plt.cm.Greens(dice / dice.max())
colors_pa = plt.cm.Reds(pa / pa.max())
colors_val_loss = plt.cm.Greys(val_loss / val_loss.max())

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# IOU
axes[0, 0].bar(models, iou, color=colors_iou)
axes[0, 0].set_title('IOU')
axes[0, 0].set_ylabel('IOU')
axes[0, 0].tick_params(axis='x', rotation=25)
add_bar_values(axes[0, 0], iou)

# Dice
axes[0, 1].bar(models, dice, color=colors_dice)
axes[0, 1].set_title('Dice')
axes[0, 1].set_ylabel('Dice')
axes[0, 1].tick_params(axis='x', rotation=25)
add_bar_values(axes[0, 1], dice)

# PA
axes[1, 0].bar(models, pa, color=colors_pa)
axes[1, 0].set_title('PA')
axes[1, 0].set_ylabel('PA')
axes[1, 0].tick_params(axis='x', rotation=25)
add_bar_values(axes[1, 0], pa)

# Validation Loss
axes[1, 1].bar(models[:-2], val_loss[:-2], color=colors_val_loss)
axes[1, 1].set_title('Validation Loss')
axes[1, 1].set_ylabel('Validation Loss')
axes[1, 1].tick_params(axis='x', rotation=25)
add_bar_values(axes[1, 1], val_loss)

plt.tight_layout()
plt.show()
```

```python
times = df['TIME']

fig, axes = plt.subplots(1, 1, figsize=(7, 5))
axes.bar(models[:-1], times[:-1], color=colors_iou)
axes.set_title('Execution Time (min)')
axes.set_ylabel('(s)')
axes.tick_params(axis='x', rotation=25)
add_bar_values(axes, times)
```
